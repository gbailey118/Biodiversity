```python
# Import libraries
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
from scipy.stats import chi2_contingency
```

```python
# load data
species = pd.read_csv('species_info.csv')
species.head()
```

```python
species.info()
```

of the 5824 records, only 191 had a non-null conservation status. Inspecting the unqiue conservation statuses, all express some level of concern therefore I have made the assumption that the null values represent species that are not of concern and have populated the field accordingly.

```python
species['conservation_status'].unique()
```

```python
species['conservation_status'] = species['conservation_status'].fillna('Not a Concern')
species.head()
```

Have created a boolean field that identifies whether the species is a conservation risk or not.

```python
species['conservation_risk'] = species['conservation_status'].apply(lambda status: True if status in ['Species of Concern', 'Endangered', 'Threatened',
       'In Recovery'] else False)
species.head()
```

```python
# load data
observations = pd.read_csv('observations.csv')
observations.head()
```

```python
observations.info()
```

```python
species.scientific_name.nunique(), observations.scientific_name.nunique()
```

The observations table has no null records. Both the species and observations table had the same number of unique species - I will use this field to merge the tables and use the combined table for my analysis.

```python
# Join Data
df = pd.merge(
    species,
    observations,
    on = 'scientific_name',
    how = 'left'
)

df.head()
```

#### 4. Chi-2 Tests

Firstly wrap the chi2_contigency function within a function that takes a single argument, a dataframe. I can then pass into this function my sliced dataframe to be able to test for significant differences between two categories of species. I am looking to test whether the conservation risk percentage is significantly different for various pairs of species.

```python
def chi2_cats(df):
    chi2, p, dof, expected = chi2_contingency(df) 

    # Results 
    print(f'Chi-Square Statistic: {chi2}') 
    print(f'p-value: {p}') 
    print(f'Degrees of Freedom: {dof}') 
    if p < 0.05:
        print("Significant")
    else: 
        print("Not Significant")
```

I took an iterative approach, testing for differences between successive categories. I started by taking mammals and birds, the two categories with the highest conservation risk percentages. The test found significant evidence that the conservation risk is not random (i.e. mammals are significantly more likely to be a conservation risk than birds).

```python
mam_bird = dfp.iloc[[1,3],]
mam_bird
```

```python
# Is there a difference between mammals and birds
chi2_cats(mam_bird)
```

The difference between birds and fish was also found to be significant.

```python
# Is there a difference between birds and fish
bird_fish = dfp.iloc[[1,2],]
chi2_cats(bird_fish)
```

The difference between fish and amphibians was not found to be significant.

```python
# Is there a difference between fish and amphibians
amp_fish = dfp.iloc[[0,2],]
chi2_cats(amp_fish)
```

The difference between fish and reptiles was not found to be significant.

```python
# Is there a difference between fish and reptiles
fish_rept = dfp.iloc[[2,5],]
chi2_cats(fish_rept)
```

The difference between reptiles and nonvascular plants was found to be significant.

```python
nv_rept = dfp.iloc[[4,5],]
chi2_cats(nv_rept)
```

The difference between vascular and nonvascular plants was found to be significant.

```python
nv_v = dfp.iloc[[4,6],]
chi2_cats(nv_v)
```
